
1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to decide whether there is enough evidence in sample data to support or reject a claim about a population.
Null Hypothesis (H₀): No effect/no difference.
Alternative Hypothesis (H₁): There is an effect/difference.
p-value vs. α (significance level):
If p ≤ α → Reject H₀.
If p > α → Fail to reject H₀.



2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- Null Hypothesis (H₀)

A default assumption or claim.
Says there is no effect, no difference, or nothing unusual.
Example: “The average exam score of students is 70.”

Alternative Hypothesis (H₁ or Ha)

What you are trying to test/prove.
Says there is an effect, a difference, or something unusual.
Example: “The average exam score of students is not 70.”

Main Difference
H₀ = status quo / baseline assumption
H₁ = research claim / new effect
In hypothesis testing, we collect data → if evidence is strong enough, we reject H₀ in favor of H₁.



3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
- Significance Level (α)
The threshold probability we set in hypothesis testing to decide whether to reject the null hypothesis (H₀).
Common values: 0.05 (5%), 0.01 (1%), 0.10 (10%).

Role in Hypothesis Testing
Compare p-value with α
If p ≤ α → Reject H₀ (evidence supports H₁).
If p > α → Fail to reject H₀ (not enough evidence against H₀).
Controls Risk of Type I Error
Type I Error = Rejecting H₀ when it is actually true (false positive).
Example: At α = 0.05, we accept a 5% chance of wrongly rejecting H₀.



4. What are Type I and Type II errors? Give examples of each.
- Type I Error (False Positive)
Definition: Rejecting the null hypothesis (H₀) when it is actually true.
Risk controlled by significance level (α).

Example:
H₀: A man is innocent.
H₁: A man is guilty.

Type I Error: Jury convicts an innocent person.
In statistics: Concluding a medicine works when in fact it doesn’t.

Type II Error (False Negative)
Definition: Failing to reject the null hypothesis when it is actually false.
Probability of this error is denoted by β (related to test power = 1 − β).

Example:
H₀: A man is innocent.
H₁: A man is guilty.

Type II Error: Jury lets a guilty person go free.
In statistics: Concluding a medicine doesn’t work when in fact it does



5. What is the difference between a Z-test and a T-test? Explain when to use
each.
- Z-Test
Used when:
Population standard deviation (σ) is known.
Sample size is large (n > 30).
Distribution: Based on the standard normal distribution (Z-distribution).

Example: Testing if the average height of 10,000 students is 170 cm when σ = 10 cm is known.

T-Test
Used when:
Population standard deviation (σ) is unknown.
Sample size is small (n ≤ 30).
Distribution: Based on the t-distribution, which is similar to normal but has fatter tails (to account for uncertainty).

Example: Testing if the average marks of a class of 20 students differ from 50, when σ is unknown.

```python
""" 7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
"""

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu0 = 50
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.4f}")
print(f"Sample Std Dev = {sample_std:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

if p_value < alpha:
    print("Reject H₀: There is significant evidence that the mean ≠ 50.")
else:
    print("Fail to Reject H₀: No significant evidence that the mean differs from 50.")
```

```python
'''8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mu, sigma = 50, 5
n = 100
data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence = 0.95
alpha = 1 - confidence
z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"Upper 95% CI = {ci_upper:.2f}")

plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
'''9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_zscores(data):

    data = np.array(data)

    mean = np.mean(data)
    std = np.std(data, ddof=1)

    z_scores = (data - mean) / std

    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_zscores(sample_data)
print("First 10 Z-scores:", z_scores[:10])
```
